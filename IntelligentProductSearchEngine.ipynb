```python
import editdistance
def is_plural(word):
    if word.endswith('s'):
        return word
    return word + 's'

def autocorrect(input_word, vocabulary):
    input_word_lemma = lemmatizer.lemmatize(input_word.lower(), pos='n')
    plural_form = is_plural(input_word_lemma)
    close_match = input_word
    dist = float('inf')
    
    for word in vocabulary:
        word_lemma = lemmatizer.lemmatize(word, pos='n')
        # Checking both singular and plural versions of the word
        for candidate in [word_lemma, is_plural(word_lemma)]:
            edit_distance = editdistance.eval(input_word_lemma, candidate)
            if edit_distance < dist:
                dist = edit_distance
                close_match = word
    
    return close_match
```

```python
import random
#Introducing random misspelling in the word
def introduce_spelling_error(word):
    operations = ['insert', 'delete', 'replace']
    operation = random.choice(operations)
    word = list(word)
    if operation == 'insert':
        position = random.randint(0, len(word))
        char = random.choice('abcdefghijklmnopqrstuvwxyz')
        word.insert(position, char)
    elif operation == 'delete' and len(word) > 1:
        position = random.randint(0, len(word) - 1)
        word.pop(position)
    elif operation == 'replace':
        position = random.randint(0, len(word) - 1)
        char = random.choice('abcdefghijklmnopqrstuvwxyz')
        word[position] = char
    return ''.join(word)





```

```python
import json

file_name = "meta_All_Beauty.jsonl"
title_words = []

with open(file_name, 'r') as fp:
    for line in fp:
        # Parse the JSON object
        record = json.loads(line.strip())
        
        # Extract the title and append the entire title string to the title_words list
        title = record.get('title', "")  # Default to an empty string if 'title' is missing

        title_words.append(title)

# Output the first 10 titles as a sample
print(title_words[:10])
```

```python
import random
sampled_titles = random.sample(title_words, 10000)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag

# Download necessary datasets
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')


noun_list = []
for title in sampled_titles:
    tokens = word_tokenize(title)
    tags = pos_tag(tokens)

    nouns = [
    word for word, tag in tags
    if tag in ('NN', 'NNS', 'NNP', 'NNPS') and word.isalpha()
    ]
    noun_list.append(nouns)
```

```python
noun_list
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import numpy as np

# Flatten your list of lists into a single list
corpus = [
    ' '.join(words) for words in noun_list
]

# Initialize the tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)
total_words = len(tokenizer.word_index) + 1

# Generate input sequences
input_sequences = []
for line in corpus:
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence)

# Pad sequences to the same length
max_sequence_len = max([len(x) for x in input_sequences])
input_sequences = pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre')

# Split inputs (X) and labels (y)
X, y = input_sequences[:,:-1], input_sequences[:,-1]
y = np.array(y)
```

```python
input_sequences
```

```python
from collections import Counter
# Creating bi-grams from the corpus
bigrams = defaultdict(list)
for sentence in corpus:
    words = sentence.split()
    for i in range(len(words)-1):
        bigrams[words[i]].append(words[i+1])



# Predicting the next-word based on frequency
def predict_next_word(word, bigrams):
    if word in bigrams:

        next_words = bigrams[word]
        word_freq = Counter(next_words)
        # Most common word
        return word_freq.most_common(1)[0][0]
    else:

        return "No prediction available at this time"

```

```python
# Correct prediciton
input_word = "razor"
predicted_word = predict_next_word(input_word, bigrams)
print(predicted_word) 
```

```python
# Incorrect prediction
input_word = "shave"
predicted_word = predict_next_word(input_word, bigrams)
print(predicted_word)  # Output expected "cream"
```

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense
from keras.utils import to_categorical


y = to_categorical(y, num_classes=total_words)

# Build the model
model = Sequential()
model.add(Embedding(input_dim=total_words, output_dim=100, input_length=max_sequence_len-1))
model.add(LSTM(150))
model.add(Dense(total_words, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=50, verbose=1)
```

```python
import json
from sentence_transformers import SentenceTransformer, util
import numpy as np

# Loading dataset
file_name = "meta_All_Beauty.jsonl"
titles = []
metadata = []

with open(file_name, 'r') as fp:
    for line in fp:
        item = json.loads(line.strip())
        title = item.get('title')
        if title:
            titles.append(title)
            metadata.append(item)  
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense
from keras.utils import to_categorical
# Load the pre-trained embedding model

# Encode all titles in the dataset
title_embeddings = model.encode(titles, convert_to_tensor=True)
```

```python
def semantic_search(query, title_embeddings, titles, metadata, model, top_n=5):
    # Embedding the query
    query_embedding = model.encode(query, convert_to_tensor=True)
    
    # Computing cosine similarities between the query and all title embeddings
    similarities = util.cos_sim(query_embedding, title_embeddings)[0]

    top_results = np.argsort(similarities.cpu().numpy())[-top_n:][::-1]
    
    # Displaying the top results with their metadata
    results = []
    for idx in top_results:
        idx = idx.item() 
        result = {
            "title": titles[idx],
            "similarity_score": similarities[idx].item(),
            "metadata": metadata[idx]
        }
        results.append(result)
    
    return results
```

```python
# Run a query
query = "brush for curly hair"
results = semantic_search(query, title_embeddings, titles, metadata, model, top_n=10)

# Display the results
for result in results:
    print("Title: "+result['title'])
    print("Similarity Score: "+str(result['similarity_score']))
    print("Metadata: "+str(result['metadata']))
    print("---------------------------------------------------")
```
